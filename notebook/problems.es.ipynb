## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python

# P(mayor a 7 o par) 
# primero voy a hacer el planteo para que tire los dados y me sume el resultado de la suma de los dos dados. hay que definir la fx 

import random
def lanza_dados():
    dado1 = random.randint(1,6)
    dado2 = random.randint(1,6)
    return dado1 + dado2


# segundo voy a hacer una variable de conteo de las veces que el resultado del tirar los dados cumpla con las dos condiciones (que me sume ahi dentro los casos fav). valor inicial 0 
conteo_cump_condiciones = 0


# hacer la funcion para que itere las 1000 veces 

iteraciones = 1000
for i in range(iteraciones):
    sum_lanza = lanza_dados()
    if sum_lanza > 7 or sum_lanza % 2 == 0:
        conteo_cump_condiciones +=1

# calcular la probabilidad con la formula basica que es el valor de veces que se han cuumplido las condiciones sobre los 1000 tiros

probabilidad = conteo_cump_condiciones/1000*100
print(f"La probabilidad de que se cumplan ambas condiciones es: {round(probabilidad, 2)}%")
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
# iteraciones 5 por cada evento que son 1000
# primero voy a hacer el planteo para que saque las bolas y me vaya colocando el resultado de la sacada en algun lugar

import random
ball_box = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
## Función para simular un evento y contar el número de bolas blancas y rojas tomadas
def simulate_event1():
    white_count = 0
    red_count = 0
    for _ in range(5):
        ball = random.choice(list(ball_box.values()))
        if ball == "White":
            white_count += 1
        elif ball == "Red":
            red_count += 1
    return white_count, red_count

# Simular el evento 1000 veces
event1_count = 0
total_simulations = 1000
for _ in range(total_simulations):
    white, red = simulate_event1()
    if white == 3 and red == 2:
        event1_count += 1

# Calcular la probabilidad del evento 1
probability_event1 = event1_count / total_simulations
print("Probabilidad de tomar 3 bolas blancas y 2 bolas rojas:", probability_event1)

# Función para simular un evento y verificar si todas las bolas son del mismo color
def simulate_event2():
    colors = set()
    for _ in range(5):
        ball = random.choice(list(ball_box.values()))
        colors.add(ball)
    return len(colors) == 1

# Simular el evento 1000 veces
event2_count = 0
for _ in range(total_simulations):
    if simulate_event2():
        event2_count += 1

# Calcular la probabilidad del evento 2
probability_event2 = event2_count / total_simulations
print("Probabilidad de tomar todas las bolas del mismo color:", probability_event2)
```
